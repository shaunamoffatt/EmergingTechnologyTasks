## Population Variance

Variance $(\sigma)^2$ is a measure of the spread of dataset[4] and can be calculated by the following steps:

    1. Calculate the mean of the dataset (ie the sum of all the values in the dataset divided by the total number of data).
    
\begin{equation}
 \bar{x} = {\frac{\sum{(x_i)}}{n}}
\end{equation}
    
    2. Individually subtracting the mean from each number in the dataset, which gives the distance each value is from the mean.
\begin{equation}
 ({x_0-\bar{x}}), ({x_1-\bar{x}}), ({x_2-\bar{x}}).... ({x_i-\bar{x}})
\end{equation}
    
    3. Squaring the result of each distance so that they are positive and adding all the positive distances together.
\begin{equation}
  {{\sum{(x-\bar{x})}^2}}
\end{equation}

    4. Finally divide the sum of the square distances by the number of values in the data set.[5]
\begin{equation}
 \sigma^2 = {\frac{\sum{(x-\bar{x})}^2}{n}}
\end{equation}



[5]https://www.sciencebuddies.org/science-fair-projects/science-fair/variance-and-standard-deviation#:~:text=You%20take%20the%20sum%20of,the%20two%20equations%20are%20equivalent.

[4] https://www150.statcan.gc.ca/n1/edu/power-pouvoir/ch12/5214891-eng.htm

## Calculate the Population Variance in the number of Incidents from the Dataset

```python
def variance(data):
    n = len(data)
    mean = sum(data)/n
    dev = [(x- mean) ** 2 for x in data]
    variance = sum(dev) /n
    return variance

meanValue = sum(incidents)/len(incidents)
print("The mean : ", meanValue)
print("The variance in the number of incidents is : ",variance(incidents))
```

### Reviewing the Variance of ~20.3

A variance of $\sim 20.3$ makes sense when looking at the graph of the data. The values range from many having 0 incidents to one even having as high a value as 24. The data is a bit spread out from the mean of 4.125, and so it's understandable that the variance would be a bit high.

## Population Standard Deviation

The Population Standard Deviation [6] ($\sigma$) is simply the positive square root of the variance $(\sigma)^2$ and is often used to measure the dispersion of data/numbers (ie scatter).[3] 

\begin{equation}
\sigma = \sqrt{\frac{\sum{(x-\bar{x})}^2}{n}}
\end{equation}

If the data set *n* is considered a "population", it means that dataset *n* contains **all** the possible data values of specific dataset.[6]

## STDEV.P

STDEV.P is the function for the Population Standard Deviation.

On the Microsoft website it describes STDEV.P using "the following formula:"[1]
\begin{equation}
\sqrt{\frac{\sum{(x-\bar{x})}^2}{n}}
\end{equation}

Which we now understand to be the Population Standard Deviation.

[1]https://support.microsoft.com/en-us/office/stdev-p-function-6e917c05-31a0-496f-ade7-4f4e7462f285

## Getting the Population Standard Deviation of the DataSet

```python
#Calculate the STDEV_P by taking the square root of the variance
def customSTDEVP(data):
    return math.sqrt(variance(data))

#Calculate the STDEV_P using numpy
def npSTDEVP(data):
    return np.sqrt(np.sum((data - np.mean(data))**2)/len(data))

incident_STDEV_P = customSTDEVP(incidents)
print("\n1. Custom function for calculation of the Population Standard Deviation of incidents is :", customSTDEVP(incidents))
print("\n2. Pythons statistic library for calculation=> statistics.pstdev(incidents) =",statistics.pstdev(incidents))
print("\n3. Finally using numpy=> np.sqrt(np.sum((incidents - np.mean(incidents))**2)/len(incidents))", npSTDEVP(incidents))
```

## Sample Variance and Standard Deviation

Sometimes it's near impossible to obtain all the data values of a population, and more often than not data scientists have to deal with samples[12]. The calculation for sample variance vs population variance is nearly identical, except for in the final step where the sum of the square distances is divided by the number of values in the data set *n* (which is the cardinality), it is instead divided by *n-1*. 

*n-1* (known as the bessel's correction[17]) results in a slightly higher variance and is used to compensate for the lack of information about the population data.[13] It allows for a more unbiased estimation.[14]

The sample standard deviation is the positive square root of the sample variance.  

\begin{equation}
s = \sqrt{\frac{\sum{(x-\bar{x})}^2}{n - 1}}
\end{equation}

In the array of incidents given, we have all the major airlines accounted for, but say we are dealing with a bigger sample set. Take for example the workforce of Ireland. Given a sample of 1 million people salarys, we are asked to estimate the standard deviantion of salarys. The work force of Ireland comprises of $\sim2.3million$ as of 2019 (according to the Central Statistics Office [15]), and if we are dealing with only 1 million salary values, our estimations will be based on $\sim43.47$% of the total workforce. Here is where the sample standard deviation would be used to get a more accurate estimation.

## Sample Standard Deviation on the random samples of incidents

Taking a few samples of our incidents array, we will calculate the standard deviation on each and compare to the population standard deviation



```python
#create a figure with 2 columns
fig, axs = plt.subplots(ncols=2,figsize=(10,5))
#add the total distr plot to one
sns.kdeplot(incidents, ax=axs[0])

list_incidents = list(incidents)
#10%,20%,40%,60% and 80%
k = []
for i in range(0,10,2):
    if i == 0:
        k.append(int(round(len(incidents)/10)))
    else:
        k.append(int(round(len(incidents)/10))*i)

# create 20 samples (2 of each percentage of the n)
sample = []
print('Each sample contains')
for j in range(5):
    sample.append(random.sample(list_incidents, k[j]))
    sns.kdeplot(sample[j], ax=axs[1])
    print('\nSample', j , sample[j])

print('\nFull Population of Incidents', list(incidents))
plt.show()
```

### Using numpy to Calculate the Standard Deviations of the sample arrays


```python
#create an array to hold the 5 sample arrays sampleStandardDev
incidents_STDEVS = []
incidents_STDEVP = []
print('Standard deviation using Bessels correction n-1')
for x in range(5):
    #Add to the array
    incidents_STDEVP.append(np.std(sample[x]))
    incidents_STDEVS.append(np.std(sample[x],ddof=1))

#Create a dataframe of the 
frame = {'Incident_STDEV.P': incidents_STDEVP, 'Incident_STDEV.S': incidents_STDEVS}
table = pd.DataFrame(frame)
table.describe()
```

## Comparing STDEV.S and STDEV.P on the 5 sample incidents

The 5 randomly taken samples of data for the incidents have quite an uneven distribution of data (e.g Delta/NorthWest airline having a high incident rate of 24 vs the majority of airlines having none or few incidents), and the results for the standard deviation are more difficult to estimate. However, it's worth noting that the mean of the *Sample* Standard Deviation **incident_STDEV.S** gets closer to the *actual* Standard Deviation of around **4.504** than using the mean of standard deviation gotten by calculating *Population* Standard deviation **incident_STDEV.P** on the 5 samples.  

# Comparing STDEV.S to STDEV.P using a random seeded sample of 100thousand

Here we will set the parameters of standard deviation and variance of random, yet normally distributed sample. 

```python
#[16]https://towardsdatascience.com/variance-sample-vs-population-3ddbd29e498a
np.random.seed(1690)
#set the variance and standard deviation
var = 7
stdev = 1
#create the normally distributed 100 thousand population sample
pop = np.random.normal(var, stdev, 100000)
#plot a 100 sample of the population
hist = plt.hist(pop, 100, density=True)
```

#### Create some tests on STDEV_P

```python
#Create arrays for holding
#samples, means, STDEVPs and STDEVSs

test_sample = []
mean = []
STDEV_P = []
STDEV_S = []

#set the number of tests to be 50
tests = 50

#loop through random samples of a size 10
#adding to the arrays
#create a figure with 2 columns
fig, axs = plt.subplots(ncols=2)
#create the distrubtion histogram for the entire population
sns.kdeplot(pop, bw=0.3, ax=axs[0])
for i in range(tests):
    sam_20 = np.random.choice(pop, 50)
    #add the second histogram of random samples of 20
    sns.kdeplot(sam_20, bw=0.3, ax=axs[1])
    test_sample.append(sam_20)
    #mean.append(np.mean(sam_20))
    STDEV_P.append(np.std(sam_20))
    STDEV_S.append(np.std(sam_20, ddof=1))

plt.show()
```

### Remarks on the distribution

```python
frame = {'STDEV.P': STDEV_P, 'STDEV.S': STDEV_S}
table = pd.DataFrame(frame)
table.describe()
```

## Conclusion of the sample of 50

Similar to results for the airline incidents samples, looking at the second row shows the mean standard deviation of both STDEV.P and STEV.S. The STDEV.S value is again closer standard deviation of 1 (that was pre-defined in this case), than the STDEV.P value.

In conclusion, STDEV.P should only be used when we are calculating standard deviation on full populations, and when given a sample of a population a more accurate result for standard deviation can be obtained by using the STDEV.S function.
